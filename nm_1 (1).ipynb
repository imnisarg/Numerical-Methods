#Q1) Forward divided difference

```python
import math
import matplotlib.pyplot as plt
def func_1(x,n):
    """
    function to generate the given exponential function
    """
    y=[0 for i in range(n+1)]
    for i in range(len(x)):
        y[i]= math.exp(-1*x[i]*x[i]/2)
    return y
def func_2(x,n):
    """
    function to generate the given sqrt function
    """
    y=[0 for i in range(n+1)]
    for i in range(len(x)):
        y[i]= math.sqrt(1+x[i]*x[i])
    return y
def func_3(x,n):
    """
    function to generate the given inverse sqrt function 
    """
    y=[0 for i in range(n+1)]
    for i in range(len(x)):
        y[i]= 1/(math.sqrt(1+(25*x[i]*x[i])))
    return y
def divided_diff(x,y,n):
    """
    function to generate forward divided diff table
    """
    table = [[0 for i in range(len(x))]for  j in range(len(x))]
    for i in range(len(x)):
        table[i][0]=y[i]
    for i in range(1,len(x)):
        
        for j in range(i,len(x)):
            table[j][i]=(table[j][i-1]-table[j-1][i-1])/(x[j]-x[j-i])
    return table
def find_new_res(res,x_pol,arr,j,i,x):
    """
    This is the function to generate the next multiplicative coefficient in the forward difference formula
    e
    """
    res*=(x_pol[i]-x[j])
    return res
def solve1(n):
    """
    This is the function to solve the function 1(exponential) by generating a forward differnce interpolation formula a
    and then plotting the results i.e y_pol,x_pol  y_orig,x_pol && error,x_pol
    
    """
    initialise_x = -1
    update_x = 2/n
    x = [0 for i in range(n+1)]
    x[0] = initialise_x
    for i in range(1,n+1):
        x[i]=x[i-1]+update_x

    y = func_1(x,n)
    table = divided_diff(x,y,n)
    arr = [0 for i in range(n+1)]
    for i in range(n+1):
        arr[i] = table[i][i]
    x_pol = [i/60 for i in range(61)]
    y_pol = [0 for i in range(61)]

    for i in range(61):
        res=x_pol[i]-x[0]
        y_pol[i] = arr[0]
        for j in range(1,n+1):
        
            y_pol[i]+= (res)*arr[j]
            res = find_new_res(res,x_pol,arr,j,i,x)
#y_pol
    y_orig=[0 for i in range(61)]
    error = [0 for i in range(61)]
    for i in range(61):
        y_orig[i] = math.exp(-1*x_pol[i]*x_pol[i]/2)
        error[i] = abs(y_orig[i]-y_pol[i])
    #import matplotlib.pyplot as plt

    fig, axes = plt.subplots(2, 2)
    #print("plot for ",n)
    axes[0, 0].plot(x_pol, y_orig,color='b')
    axes[0,0].set_title("x and original y")
    axes[0, 1].plot(x_pol, y_pol,color='r')
    axes[0,1].set_title("x and y through polynomial")
    axes[1,1].plot(x_pol,error,color='g')
    axes[1,1].set_title("error")
def solve2(n):
    """
    This is the function to solve the function 2(square root) by generating a forward differnce interpolation formula a
    and then plotting the results i.e y_pol,x_pol  y_orig,x_pol && error,x_pol

    """
    initialise_x = -1
    update_x = 2/n
    x = [0 for i in range(n+1)]
    x[0] = initialise_x
    for i in range(1,n+1):
        x[i]=x[i-1]+update_x

    y = func_2(x,n)
    table = divided_diff(x,y,n)
    arr = [0 for i in range(n+1)]
    for i in range(n+1):
        arr[i] = table[i][i]
    x_pol = [i/60 for i in range(61)]
    y_pol = [0 for i in range(61)]

    for i in range(61):
        res=x_pol[i]-x[0]
        y_pol[i] = arr[0]
        for j in range(1,n+1):
        
            y_pol[i]+= (res)*arr[j]
            res = find_new_res(res,x_pol,arr,j,i,x)
#y_pol
    y_orig=[0 for i in range(61)]
    error = [0 for i in range(61)]
    for i in range(61):
        y_orig[i] = math.sqrt(1+x_pol[i]*x_pol[i])
        error[i] = abs(y_orig[i]-y_pol[i])
    #import matplotlib.pyplot as plt

    fig, axes = plt.subplots(2, 2)
    #print("plot for ",n)
    axes[0, 0].plot(x_pol, y_orig,color='b')
    axes[0,0].set_title("x and original y")
    axes[0, 1].plot(x_pol, y_pol,color='r')
    axes[0,1].set_title("x and y through polynomial")
    axes[1,1].plot(x_pol,error,color='g')
    axes[1,1].set_title("error")
def solve3(n):
    """
    This is the function to solve the function 3(inverse square root) by generating a forward differnce interpolation formula a
    and then plotting the results i.e y_pol,x_pol  y_orig,x_pol && error,x_pol
    
    """
    initialise_x = -1
    update_x = 2/n
    x = [0 for i in range(n+1)]
    x[0] = initialise_x
    for i in range(1,n+1):
        x[i]=x[i-1]+update_x

    y = func_3(x,n)
    table = divided_diff(x,y,n)
    arr = [0 for i in range(n+1)]
    for i in range(n+1):
        arr[i] = table[i][i]
    x_pol = [i/60 for i in range(61)]
    y_pol = [0 for i in range(61)]

    for i in range(61):
        res=x_pol[i]-x[0]
        y_pol[i] = arr[0]
        for j in range(1,n+1):
        
            y_pol[i]+= (res)*arr[j]
            res = find_new_res(res,x_pol,arr,j,i,x)
#y_pol
    y_orig=[0 for i in range(61)]
    error = [0 for i in range(61)]
    for i in range(61):
        y_orig[i] = 1/(math.sqrt(1+(25*x_pol[i]*x_pol[i])))
        error[i] = abs(y_orig[i]-y_pol[i])
    #import matplotlib.pyplot as plt

    fig, axes = plt.subplots(2, 2)
    #print("plot for ",n)
    axes[0, 0].plot(x_pol, y_orig,color='b')
    axes[0,0].set_title("x and original y")
    axes[0, 1].plot(x_pol, y_pol,color='r')
    axes[0,1].set_title("x and y through polynomial")
    axes[1,1].plot(x_pol,error,color='g')
    axes[1,1].set_title("error")
```
